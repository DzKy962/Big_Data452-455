# Pra-pemrosesan Data

Pre-processing adalah proses membersihkan mempersiapkan dataset untuk dilakukanya analisis dan pemodelan, proses ini sangat penting untuk menghindari kesalahan analisis, error, null data(data yang hilang) maupun data yang rusak. Dalam Penelitian ini, Kami melakukan beberapa pre-processing seperti Data Cleaning, Data Tranformation dan Rekayasa Fitur yang dilakukan dari 2 Fitur Yaitu (Valence dan Energy).

##### a. Bersihkan data

```python
missing_values = df.isnull().sum() # Memeriksa missing value di seluruh DataFrame
print(missing_values)
```

```python
# Menangani nilai yang hilang, data duplikat, dan outlier
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
```

```python
missing_values = df.isnull().sum() # Memeriksa missing value di seluruh DataFrame
print(missing_values)
```

##### b. Transformasi data

```python
# untuk menormalkan fitur audio 'loudness' dalam dataset
def normalize(series):
    return (series - series.min()) / (series.max() - series.min())

df['normalized_loudness'] = normalize(df['loudness'])
```

##### c. Rekayasa fitur

```python
# untuk membuat fitur baru yang disebut 'mood_index' dengan menggabungkan dua fitur audio yang sudah ada, yaitu 'valence' dan 'energy'
df['mood_index'] = (df['valence'] + df['energy']) / 2
```

# Analisis Data Eksplorasi (EDA)

Sesuai dengan nama prosesnya, EDA atau Eloratory Data Analysis adalah proses yang dilakukan untuk menggali, mencari, mengungkap wawasan dan informasi dari dataset yang akan digunakan dalama penelitian tersebut. Informasi yang didapatkan akan sangat krusial dan penting dalam pengembangan penelitian yang dilakukan. Beberapa proses EDA yang dilakukan disini adalah
*   Visualisasi distribusi fitur
*   Identifikasi korelasi dengan matrix
*   Analisis playlist
*   Segmentasi audiens dengan Elbow
*   Tren dari waktu ke waktu
*   Statistik deskriptif

##### a. Visualisasi distribusi fitur

```python
plt.figure(figsize=(10, 6)) # Membuat figure dengan ukuran 10x6 inci.
sns.histplot(df['danceability'], kde=True, bins=30, color='blue', label='Danceability') # Membuat histogram untuk 'danceability' dengan KDE, 30 bins, warna biru, dan label.
sns.histplot(df['energy'], kde=True, bins=30, color='orange', label='Energy') # Membuat histogram untuk 'energy' dengan KDE, 30 bins, warna oranye, dan label.
plt.title('Distribution of Audio Features') # Menambahkan judul plot.
plt.legend() # Menampilkan legenda.
plt.show() # Menampilkan plot.
```

##### b. Identifikasi korelasi dengan matrix

```python
correlation_matrix = df[['danceability', 'energy', 'loudness', 'valence', 'tempo']].corr()  # Menghitung matriks korelasi untuk fitur yang dipilih.
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')  # Membuat heatmap dari matriks korelasi dengan anotasi dan colormap 'coolwarm'.
plt.title('Correlation Matrix of Audio Features')  # Menambahkan judul plot.
plt.show()  # Menampilkan plot.
```

##### c. Analisis playlist

```python
playlist_analysis = df.groupby('playlist_genre')[['danceability', 'energy', 'valence']].mean() # Menghitung rata-rata danceability, energy, dan valence untuk setiap genre playlist.
sns.set_style("whitegrid") # Mengatur gaya plot Seaborn ke "whitegrid".
playlist_analysis.plot(kind='bar', figsize=(12, 6), width=0.8, color=['skyblue', 'coral', 'lightgreen']) # Membuat diagram batang dengan ukuran, lebar batang, dan warna yang ditentukan.
plt.title('Rata-rata Danceability, Energy, dan Valence per Genre Playlist', fontsize=16) # Menambahkan judul plot dengan ukuran font.
plt.xlabel('Genre Playlist', fontsize=12) # Menambahkan label sumbu x dengan ukuran font.
plt.ylabel('Nilai Rata-rata', fontsize=12) # Menambahkan label sumbu y dengan ukuran font.
plt.xticks(rotation=45, ha='right', fontsize=10) # Memutar label sumbu x 45 derajat, rata kanan, dengan ukuran font.
plt.legend(fontsize=10) # Menampilkan legenda dengan ukuran font.
plt.tight_layout() # Menyesuaikan tata letak plot agar tidak tumpang tindih.
plt.show() # Menampilkan plot.
```

##### d. Segmentasi audiens dengan Elbow

```python
kmeans = KMeans(n_clusters=3, random_state=42) # Inisialisasi model KMeans dengan 3 cluster dan random state 42.
df['audience_segment'] = kmeans.fit_predict(df[['danceability', 'energy', 'valence']]) # Melakukan clustering dan menambahkan kolom 'audience_segment' ke DataFrame.
print(df[['playlist_name', 'audience_segment']].head(10)) # Menampilkan 10 playlist_name dan audience_segment teratas.
```

```python
inertia_values = [] # Inisialisasi list untuk menyimpan nilai inertia.
for num_clusters in range(1, 11): # Loop untuk jumlah cluster dari 1 hingga 10.
    kmeans = KMeans(n_clusters=num_clusters, random_state=42) # Inisialisasi model KMeans dengan jumlah cluster dan random state.
    kmeans.fit(df[['danceability', 'energy', 'valence']]) # Melatih model KMeans dengan fitur yang relevan.
    inertia_values.append(kmeans.inertia_) # Menyimpan nilai inertia ke dalam list.
plt.figure(figsize=(8, 6)) # Membuat figure dengan ukuran 8x6 inci.
plt.plot(range(1, 11), inertia_values, marker='o', label='Inertia') # Membuat plot inertia terhadap jumlah cluster.
plt.axvline(x=3, color='red', linestyle='--', label='Letak Cluster Optimal: 3') # Menambahkan garis vertikal untuk menandai cluster optimal.
plt.title('Elbow Method untuk Menentukan Jumlah Cluster Optimal') # Menambahkan judul plot.
plt.xlabel('Jumlah Cluster') # Menambahkan label sumbu x.
plt.ylabel('Inertia') # Menambahkan label sumbu y.
plt.legend(loc='best') # Menampilkan legenda pada posisi terbaik.
plt.grid(alpha=0.3) # Menambahkan grid dengan transparansi.
plt.show() # Menampilkan plot.
```

##### e. Tren dari waktu ke waktu

```python
df['track_album_release_date'] = pd.to_datetime(df['track_album_release_date'], errors='coerce', format='%Y-%m-%d') # Mengubah kolom 'track_album_release_date' ke tipe data datetime, menangani kesalahan dengan 'coerce', dan menentukan format '%Y-%m-%d'.
df['release_year'] = df['track_album_release_date'].dt.year # Membuat kolom baru 'release_year' dengan mengekstrak tahun dari kolom 'track_album_release_date'.
```

```python
# Menghitung rata-rata popularitas lagu per 5 tahun
print("\nMenghitung rata-rata popularitas lagu per 5 tahun:\n") # Menampilkan judul

popularity_by_5year = df.groupby(pd.cut(df['release_year'], bins=range(int(df['release_year'].min()), int(df['release_year'].max()) + 6, 5)), observed=False)['track_popularity'].mean() # Mengelompokkan data berdasarkan rentang 5 tahun rilis dan menghitung rata-rata popularitas lagu untuk setiap rentang.

# Plot data
plt.figure(figsize=(12, 6))  # Membuat figure dengan ukuran 12x6 inci.
plt.plot(popularity_by_5year.index.astype(str), popularity_by_5year.values, marker='o')  # Membuat plot garis dengan index (rentang tahun) sebagai sumbu x dan rata-rata popularitas sebagai sumbu y, menggunakan marker 'o'.
plt.title('Rata-rata Popularitas Lagu berdasarkan Rentang 5 Tahun Rilis')  # Menambahkan judul plot.
plt.xlabel('Rentang Tahun Rilis')  # Menambahkan label sumbu x.
plt.ylabel('Rata-rata Popularitas')  # Menambahkan label sumbu y.
plt.xticks(rotation=45, ha='right')  # Memutar label sumbu x 45 derajat ke kanan agar mudah dibaca.
plt.grid(True)  # Menambahkan grid ke plot.
plt.show()  # Menampilkan plot.
```

```python
# Distribusi Fitur Audio berdasarkan Tahun Rilis (5-year intervals)
print("\nDistribusi Fitur Audio berdasarkan Tahun Rilis Setiap 5 tahun:\n") # Menampilkan judul

# Fitur audio yang ingin divisualisasikan
features = ['danceability', 'energy', 'valence']  # Mendefinisikan fitur audio yang akan dianalisis.

# Menghitung rata-rata fitur audio per 5 tahun rilis
feature_by_5year = df.groupby(pd.cut(df['release_year'], bins=range(int(df['release_year'].min()), int(df['release_year'].max()) + 6, 5)), observed=False)[features].mean().reset_index() # Mengelompokkan data berdasarkan rentang 5 tahun rilis dan menghitung rata-rata fitur audio untuk setiap rentang.

# Mengubah index menjadi string untuk visualisasi yang lebih baik
feature_by_5year['release_year'] = feature_by_5year['release_year'].astype(str)  # Mengubah tipe data kolom 'release_year' menjadi string agar dapat digunakan sebagai label sumbu x.

# Membuat diagram batang
fig, axes = plt.subplots(nrows=len(features), ncols=1, figsize=(12, 6 * len(features)))  # Membuat subplot untuk setiap fitur audio dengan ukuran yang ditentukan.

for i, feature in enumerate(features):  # Melakukan iterasi melalui setiap fitur audio.
    sns.barplot(x='release_year', y=feature, data=feature_by_5year, ax=axes[i], color='skyblue')  # Membuat bar plot untuk setiap fitur audio dengan rentang tahun rilis pada sumbu x dan nilai rata-rata fitur pada sumbu y.
    axes[i].set_title(f'Rata-rata {feature} berdasarkan Rentang 5 Tahun Rilis', fontsize=14)  # Menambahkan judul ke setiap subplot.
    axes[i].set_xlabel('Rentang Tahun Rilis', fontsize=12)  # Menambahkan label sumbu x.
    axes[i].set_ylabel(feature, fontsize=12)  # Menambahkan label sumbu y.
    axes[i].tick_params(axis='x', rotation=45, labelsize=10)  # Memutar label sumbu x 45 derajat agar mudah dibaca.

plt.tight_layout()  # Menyesuaikan tata letak subplot agar tidak tumpang tindih.
plt.show()  # Menampilkan plot.
```

```python
# Korelasi antara tahun rilis dan popularitas lagu
print("\nKorelasi antara tahun rilis dan popularitas lagu di setiap 5 tahun:\n") # Menampilkan judul

# Mendefinisikan level popularitas
popularity_levels = ['Low (0-49)', 'Medium (50-79)', 'High (80-100)']

# Mengkategorikan lagu berdasarkan popularitas
df['popularity_level'] = pd.cut(df['track_popularity'], bins=[0, 49, 79, 100], labels=popularity_levels) # Membuat kolom baru 'popularity_level' berdasarkan rentang popularitas.

# Mengelompokkan berdasarkan tahun rilis dan level popularitas, lalu menghitung jumlah lagu
popularity_by_year = df.groupby(['release_year', 'popularity_level'], observed=False)['track_id'].count().reset_index() # Mengelompokkan data berdasarkan tahun dan level popularitas, menghitung jumlah lagu, dan mereset index.

# Memutar tabel untuk memudahkan plotting
popularity_by_year = popularity_by_year.pivot(index='release_year', columns='popularity_level', values='track_id').fillna(0) # Memutar tabel agar tahun rilis menjadi index, level popularitas menjadi kolom, dan jumlah lagu menjadi nilai.

# Mengelompokkan berdasarkan interval 5 tahun untuk tampilan yang lebih jelas
popularity_by_5year = df.groupby([pd.cut(df['release_year'], bins=range(int(df['release_year'].min()), int(df['release_year'].max()) + 6, 5)), 'popularity_level'], observed=False)['track_id'].count().reset_index() # Mengelompokkan data berdasarkan interval 5 tahun dan level popularitas, menghitung jumlah lagu, dan mereset index.
popularity_by_5year = popularity_by_5year.pivot(index='release_year', columns='popularity_level', values='track_id').fillna(0) # Memutar tabel agar interval 5 tahun menjadi index, level popularitas menjadi kolom, dan jumlah lagu menjadi nilai.

# Membuat plot batang bertumpuk
popularity_by_5year.plot(kind='bar', stacked=True, figsize=(12, 6), color=['lightcoral', 'lightskyblue', 'lightgreen']) # Membuat plot batang bertumpuk dengan ukuran dan warna yang ditentukan.
plt.title('Distribution of Song Popularity Levels by 5-Year Release Intervals', fontsize=14) # Menambahkan judul plot.
plt.xlabel('Release Year Interval', fontsize=12) # Menambahkan label sumbu x.
plt.ylabel('Number of Songs', fontsize=12) # Menambahkan label sumbu y.
plt.xticks(rotation=45, ha='right', fontsize=10) # Memutar label sumbu x 45 derajat ke kanan agar mudah dibaca.
plt.legend(title='Popularity Levels', loc='best') # Menampilkan legenda pada posisi terbaik.
plt.tight_layout() # Menyesuaikan tata letak plot agar tidak tumpang tindih.
plt.show() # Menampilkan plot.
```

# Optimalisasi Strategi Musik Digital

##### a. Identifikasi karakteristik lagu populer

```python
popular_tracks = df[df['track_popularity'] >= 60] # Memfilter data untuk mendapatkan lagu-lagu yang dianggap populer (track_popularity >= 60).
popular_features = popular_tracks[['danceability', 'energy', 'valence']].mean() # Menghitung nilai rata-rata untuk fitur 'danceability', 'energy', dan 'valence' dari lagu-lagu populer.
print('Karakteristik Lagu Populer:', popular_features) # Menampilkan hasil rata-rata fitur lagu populer.
```

##### b. Optimasi pembuatan musik
Wawasan dari analisis digunakan untuk membuat panduan pembuatan musik

```python
print('\nRekomendasi rentang nilai untuk fitur audio berdasarkan lagu populer:') # Menampilkan judul.
for feature, value in popular_features.items(): # Melakukan iterasi melalui setiap fitur (danceability, energy, valence) dan nilai rata-ratanya.
    # Mencari nilai minimum dan maksimum untuk setiap fitur pada lagu populer
    min_value = df[df['track_popularity'] >= 60][feature].min() # Mendapatkan nilai minimum untuk fitur saat ini dari lagu-lagu populer.
    max_value = df[df['track_popularity'] >= 60][feature].max() # Mendapatkan nilai maksimum untuk fitur saat ini dari lagu-lagu populer.

    print(f'- {feature}: targetkan nilai antara {min_value:.2f} dan {max_value:.2f} (nilai rata-rata: {value:.2f})') # Menampilkan rekomendasi rentang nilai untuk fitur saat ini.
```

##### c. Personalisasi rekomendasi
Model prediktif digunakan untuk memberikan rekomendasi lagu kepada pengguna

Fungsi untuk merekomendasikan lagu berdasarkan preferensi pengguna

```python
from IPython.display import display, HTML
def rekomendasi_lagu(preferensi_pengguna, num_recommendations=5):  # Mendefinisikan fungsi 'rekomendasi_lagu' dengan parameter preferensi pengguna dan jumlah rekomendasi.
    """
    Memberikan rekomendasi lagu berdasarkan preferensi pengguna.

    Args:
        preferensi_pengguna (dict): Kamus berisi preferensi pengguna untuk fitur audio (misalnya, {'danceability': 0.7, 'energy': 0.8}).
        num_recommendations (int): Jumlah rekomendasi yang ingin diberikan (default: 5).

    Returns:
        pandas.DataFrame: Tabel berisi lagu yang direkomendasikan.
    """
    # Mencari lagu yang sesuai dengan preferensi pengguna
    rekomendasi = df[  # Memfilter DataFrame 'df' untuk menemukan lagu yang cocok.
        (df['danceability'] >= preferensi_pengguna['danceability'] - 0.1) &  # Memeriksa apakah 'danceability' lagu dalam rentang preferensi pengguna.
        (df['danceability'] <= preferensi_pengguna['danceability'] + 0.1) &  # Memeriksa apakah 'danceability' lagu dalam rentang preferensi pengguna.
        (df['energy'] >= preferensi_pengguna['energy'] - 0.1) &  # Memeriksa apakah 'energy' lagu dalam rentang preferensi pengguna.
        (df['energy'] <= preferensi_pengguna['energy'] + 0.1)  # Memeriksa apakah 'energy' lagu dalam rentang preferensi pengguna.
    ][['track_name', 'track_artist', 'danceability', 'energy']].head(num_recommendations)  # Memilih kolom yang relevan dan membatasi jumlah rekomendasi.

    return rekomendasi  # Mengembalikan DataFrame berisi lagu yang direkomendasikan.

# Tempat penggunaan personalisasi:
preferensi = {'danceability': 0.98, 'energy': 1.0}  # Mendefinisikan preferensi pengguna.
rekomendasi = rekomendasi_lagu(preferensi, num_recommendations=10)  # Memanggil fungsi 'rekomendasi_lagu' untuk mendapatkan rekomendasi.

print(f'\nRekomendasi Lagu untuk preferensi {preferensi}:\n')  # Menampilkan judul rekomendasi.
display(HTML(rekomendasi.to_html(index=False))) # Menampilkan tabel dalam format HTML, index=False untuk menyembunyikan index kolom
```
